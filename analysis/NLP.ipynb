```python
import spacy
from tqdm import tqdm
from langdetect import detect
from pathos.multiprocessing import ProcessingPool as Pool
import pandas as pd

# Load SpaCy models and disable unnecessary components
spacy_models = {
    'en': spacy.load('en_core_web_sm', disable=['ner', 'parser']),
    'fr': spacy.load('fr_core_news_sm', disable=['ner', 'parser']),
    'es': spacy.load('es_core_web_sm', disable=['ner', 'parser']),
    'de': spacy.load('de_core_web_sm', disable=['ner', 'parser']),
    'pl': spacy.load('pl_core_news_sm', disable=['ner', 'parser']),
}

# Function to process a batch of texts for lemmatization
def batch_lemmatize_multilingual(texts):
    results = []
    for text in texts:
        try:
            lang = detect(text)  # Detect language
            if lang in spacy_models:
                nlp = spacy_models[lang]
                doc = nlp(text)
                results.append(' '.join([token.lemma_ for token in doc]))
            else:
                results.append(text)  # Return as-is if the language is unsupported
        except Exception:
            results.append(text)  # Return as-is in case of an error
    return results

# Function to split DataFrame into chunks
def split_dataframe(df, chunk_size):
    for i in range(0, len(df), chunk_size):
        yield df[i:i + chunk_size]

# Parallel processing function for Jupyter
def parallel_process_lemmatization(df, column, chunk_size, num_workers):
    chunks = list(split_dataframe(df[column], chunk_size))
    tqdm_total = len(df)  # Total number of entries for progress tracking

    # Use pathos for multiprocessing
    pool = Pool(num_workers)
    results = []
    for result in tqdm(pool.imap(batch_lemmatize_multilingual, chunks), total=len(chunks), desc="Processing Batches"):
        results.extend(result)

    pool.close()
    pool.join()
    return results

# Data example
df_ratings_text_example = pd.DataFrame({
    'text_cleaned_no_stop_words': [
                                      "This is an example sentence in English.",
                                      "Ceci est une phrase en français.",
                                      "Esta es una oración en español.",
                                      "Dies ist ein Satz auf Deutsch.",
                                      "To jest zdanie po polsku."
                                  ] * 1400  # Simulating a large dataset
})

# Parameters
chunk_size = 1000  # Adjust chunk size based on memory
num_workers = 4  # Set number of processes

# Process with progress bar
tqdm.pandas(desc="Overall Progress")
df_ratings_text_example['text_cleaned_lemma'] = parallel_process_lemmatization(
    df_ratings_text_example,
    'text_cleaned_no_stop_words',
    chunk_size=chunk_size,
    num_workers=num_workers
)

# Output first few rows to verify
df_ratings_text_example.head()
```

```python
# nltk.download('wordnet')

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    st = ""
    for w in w_tokenizer.tokenize(text):
        st = st + lemmatizer.lemmatize(w) + " "
    return st.strip()

df_ratings_text['text_cleaned_lemma'] = df_ratings_text['text_cleaned_no_stop_words'].apply(lemmatize_text)
```

```python
df['tokens_flat'] = df['tokens'].apply(lambda x: [item for sublist in x for item in sublist])
```

```python
import pandas as pd
from collections import Counter

histograms = df[['country_user', 'tokens_flat']].copy()
histograms['word_histogram'] = histograms['tokens_flat'].apply(lambda x: Counter(x))

# Convert the histograms to a readable DataFrame format
histogram_df = pd.DataFrame([
    {'country': row['country_user'], 'word': word, 'count': count}
    for _, row in histograms.iterrows()
    for word, count in row['word_histogram'].items()
])
```

```python
histogram_df.head()
```

```python
# sort the df by country then by count
histogram_df = histogram_df.sort_values(by=['country', 'count'], ascending=False)
```

```python
histogram_df.head()
```

```python
# show the top words for each country
countries = histogram_df['country'].unique()
top_words_per_country = pd.DataFrame(columns=['country', 'word', 'count'])
for country in countries:
    top_words_per_country = pd.concat([top_words_per_country, histogram_df[histogram_df['country'] == country].head(10)])
```

```python
# for each country take the top 1 word (unless it's empty: '' or ' ') and drop duplicates
top_words_per_country = top_words_per_country.where(top_words_per_country['word'] != '').where(top_words_per_country['word'] != ' ').groupby('country').head(3)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python

# Create a wordmap using networkx
G = nx.Graph()

# Add countries and words as nodes
countries = top_words_per_country['country'].unique()
words = top_words_per_country['word'].unique()
G.add_nodes_from(countries, type='country')
G.add_nodes_from(words, type='word')

# Add edges with weights based on count
for _, row in top_words_per_country.iterrows():
    G.add_edge(row['country'], row['word'], weight=row['count'])

# Plot the graph
plt.figure(figsize=(12, 12))

# Generate positions for nodes
pos = nx.spring_layout(G, seed=42)

# Draw nodes with different colors based on type
country_nodes = [node for node, attr in G.nodes(data=True) if attr['type'] == 'country']
word_nodes = [node for node, attr in G.nodes(data=True) if attr['type'] == 'word']

nx.draw_networkx_nodes(G, pos, nodelist=country_nodes, node_color='blue', label='Countries', node_size=1000)
nx.draw_networkx_nodes(G, pos, nodelist=word_nodes, node_color='orange', label='Words', node_size=700)

# Draw edges with transparency based on weight
edges = G.edges(data=True)
nx.draw_networkx_edges(G, pos, edgelist=edges, alpha=0.6, width=[edge[2]['weight'] / 5 for edge in edges])

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=8)

# Add legend and title
plt.legend(scatterpoints=1, loc='upper right')
plt.title("Wordmap of Countries and Their Associated Words")
plt.show()
```

```python
# Group and aggregate data to create a more compact and meaningful visualization
import matplotlib.colors as mcolors

# Aggregate the counts for countries and words
aggregated_data = top_words_per_country.groupby(['country', 'word']).sum().reset_index()

# Create a pivot table for a heatmap visualization
heatmap_data = aggregated_data.pivot(index='country', columns='word', values='count').fillna(0)

# Plot a heatmap
plt.figure(figsize=(12, 8))
plt.imshow(heatmap_data, cmap='YlGnBu', aspect='auto')

# Add labels for countries and words
plt.xticks(ticks=range(heatmap_data.shape[1]), labels=heatmap_data.columns, rotation=45, ha='right')
plt.yticks(ticks=range(heatmap_data.shape[0]), labels=heatmap_data.index)

# Add colorbar and title
plt.colorbar(label='Word Count')
plt.title('Heatmap of Word Associations by Country')
plt.xlabel('Words')
plt.ylabel('Countries')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Define a function to dynamically bin the word counts into categories
def bin_counts(count):
    if count < 10:
        return 'Low'
    elif 10 <= count < 100:
        return 'Medium'
    elif 100 <= count < 1000:
        return 'High'
    else:
        return 'Very High'

# Apply dynamic binning to the DataFrame
top_words_per_country['count_bin'] = top_words_per_country['count'].apply(bin_counts)

# Create a pivot table for binned counts
binned_heatmap_data = top_words_per_country.pivot_table(index='country', columns='count_bin', values='count', aggfunc='sum').fillna(0)

# Plot a heatmap with the binned data
plt.figure(figsize=(12, 8))
plt.imshow(binned_heatmap_data, cmap='YlGnBu', aspect='auto')

# Add labels for countries and bins
plt.xticks(ticks=range(binned_heatmap_data.shape[1]), labels=binned_heatmap_data.columns, rotation=45, ha='right')
plt.yticks(ticks=range(binned_heatmap_data.shape[0]), labels=binned_heatmap_data.index)

# Add colorbar and title
plt.colorbar(label='Binned Word Count')
plt.title('Binned Heatmap of Word Associations by Country')
plt.xlabel('Count Categories')
plt.ylabel('Countries')
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import requests
import tempfile
import zipfile
import os

def download_world_map():
    """Download and load the Natural Earth world map data."""
    # URL for Natural Earth 1:110m Cultural Vectors
    url = "https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/110m/cultural/ne_110m_admin_0_countries.zip"

    # Create a temporary directory
    temp_dir = tempfile.mkdtemp()

    try:
        # Download the zip file
        response = requests.get(url)
        zip_path = os.path.join(temp_dir, "world_map.zip")

        with open(zip_path, 'wb') as f:
            f.write(response.content)

        # Extract the zip file
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(temp_dir)

        # Load the shapefile
        shapefile_path = os.path.join(temp_dir, "ne_110m_admin_0_countries.shp")
        world = gpd.read_file(shapefile_path)

        return world

    except Exception as e:
        print(f"Error downloading or processing world map data: {e}")
        return None

def create_word_map(top_words_df):
    # Load world map data
    world = download_world_map()

    if world is None:
        raise Exception("Failed to load world map data")

    # Merge the word data with the map data
    # Note: Using 'NAME' instead of 'name' as that's the column name in the Natural Earth data
    world = world.merge(top_words_df, how='left', left_on=['NAME'], right_on=['country'])

    # Create the plot
    fig, ax = plt.subplots(figsize=(20, 10))

    # Plot the base map
    world.plot(ax=ax, color='lightgray')

    # Plot countries with data in a different color
    world[~world['word'].isna()].plot(ax=ax, color='lightblue')

    # Add word labels
    for idx, row in world[~world['word'].isna()].iterrows():
        # Get the center point of each country
        centroid = row.geometry.centroid
        ax.annotate(
            row['word'],
            xy=(centroid.x, centroid.y),
            xytext=(0, 0),
            textcoords="offset points",
            ha='center',
            va='center',
            fontsize=8
        )

    # Customize the map
    ax.set_title('Most Common Word by Country', fontsize=16)
    ax.axis('off')

    # Add a legend for countries with/without data
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='lightblue', label='Has data'),
        Patch(facecolor='lightgray', label='No data')
    ]
    ax.legend(handles=legend_elements, loc='lower left')

    return fig
```
